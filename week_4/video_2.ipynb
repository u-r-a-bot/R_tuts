## Measures of Shape
Using library tseries and moments

```r
library(tseries)
library(moments)
Norm = rnorm(10000)
skewness(Norm)
skewness(data$Heart.Rate)
```

```r
agostino.test(Norm)
```

### Inference from above cell
* P-value is greater than 0.05, which means that null hypothesis is not rejected considering **5%** level of significance.

```r
agostino.test(data$Heart.Rate)
```

### Inference from above cell
* P-value is less than 0.05, which means that null hypothesis is rejected considering **5%** level of significance.
