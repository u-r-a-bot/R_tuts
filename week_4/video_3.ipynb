## Cluster Sampling
* Using Library Sampling

```r
library(sampling)
data$Sex = as.factor(round(runif(8763,1,50),0))
data_sample = cluster(data , clustername = c('Sex'), size= 10 , method = 'srswor')
dim(data_sample)
```

```r
levels(data_sample$Sex)
```

```r
data_sample$Sex = droplevels(data_sample$Sex)
```

```r
levels(data_sample)
```

```r
Final_sample = data_sample %>%group_by(Sex)%>% sample_n(size=5)
dim(Final_sample)
```

```r
New_data = getdata(data , Final_sample)
summary(New_data$Heart.Rate)
```

## Confidence Interval Estimation

```r
library(distributions3)
mean(data$Heart.Rate)
```

```r
a=0.05
1-a
```

```r
Norm = Normal(0,1)
Norm$mu
```

```r
plot(Norm)
```

```r
SE = sd(data$Heart.Rate) / sqrt(length(data$Heart.Rate))
z = quantile(Norm , 1-a/2)
```

```r
MOE = z*SE
lower = mean(Heart_rate) - MOE
upper = mean(Heart_rate) + MOE
```

```r
mean(Heart_rate)
```

## Confidence Interval estimation using t-Statistic

```r
DOF = length(Heart_rate-1)
t = qt(p=a/2 , df = DOF,lower.tail = F )
SE = sd(Heart_rate)/sqrt(length(Heart_rate))
```

```r
MOE = t*SE
```

```r
lower = mean(Heart_rate) - MOE
upper = mean(Heart_rate) - MOE
```
