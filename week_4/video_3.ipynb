##  Range Measures and dispersion measures

```r
var(data$Heart.Rate)
var(data$Cholesterol)
```

```r
sd(data$Heart.Rate)
sd(data$Cholesterol)
```

## Visualization

```r
plot(density(data$Heart.Rate), main= 'Density-Graph', xlab = 'Data', col='red', lwd = 4)
```

```r
Norm_pre  = rnorm(1000, mean=mean(data$Heart.Rate) , sd=sd(data$Heart.Rate))
plot(density(data$Heart.Rate), main= 'Density-Graph', xlab = 'Data', col='red', lwd = 4)
lines(density(Norm_pre), col='green', lwd=4)
```
